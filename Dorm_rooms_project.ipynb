```python
class Student:
  def __init__(self, ruid, max_price, preferred_hall_ids, roommate_preferences, accessibility_need, preferred_residence_type):
    self.ruid = ruid
    self.max_price = max_price
    self.preferred_hall_ids = preferred_hall_ids
    self.roommate_preferences = roommate_preferences
    self.accessibility_need = accessibility_need
    self.preferred_residence_type = preferred_residence_type

student_objects = []

for index, row in students_df_A.iterrows():
    student = Student(
        ruid=row['RUID'],
        max_price=row['max_price'],
        preferred_hall_ids=row['preferred_hall_ids'].strip("[]").replace("'", "").split(),
        roommate_preferences=row['roommate_preferences'].strip(),
        accessibility_need=row['accessibility_need'],
        preferred_residence_type=row['preferred_residence_type']
    )
    student_objects.append(student)

#for student in student_objects[:5]:
#    print(f"Student RUID: {student.ruid}, Max Price: {student.max_price}, Preferred Halls: {student.preferred_hall_ids}, Roommate Prefs: {student.roommate_preferences}, Accessibility: {student.accessibility_need}, Residence Type: {student.preferred_residence_type}")


class Room:
  def __init__(self, room_id, hall_id, price, is_accessible, residence_type, floor_plan, room_type):
    self.room_id = room_id
    self.hall_id = hall_id
    self.price = price
    self.is_accessible = is_accessible
    self.residence_type = residence_type
    self.floor_plan = floor_plan
    self.room_type = room_type

room_objects = []

for index, row in rooms_df.iterrows():
    room = Room(
        room_id=row['room_id'],
        hall_id=row['hall_id'],
        price=row['price'],
        is_accessible=row['has_accessibility_ramps'],
        residence_type=row['residence_type']
    )
    room_objects.append(room)

# for room in room_objects[:5]:
#   print(room.room_id, room.hall_id, room.price, room.is_accessible, room.residence_type)


hall_assignments = {room.room_id: [] for room in room_objects}
print(hall_assignments)


def match_students_to_rooms():
    for student in student_objects:
        for preferred_hall in student.preferred_hall_ids:
            for room in room_objects:
                if room.hall_id == preferred_hall and room.assigned_student is None:
                    room.assigned_student = student.ruid
                    hall_assignments[room.room_id].append(student.ruid)
                    print(f"Assigned Student {student.ruid} to Room {room.room_id} in Hall {preferred_hall}")
                    break
            else:
                continue
            break

match_students_to_rooms()





# for student in student_objects[:5]:
#     print(f"Student RUID: {student.ruid} (Type: {type(student.ruid)})")
#     print(f"Max Price: {student.max_price} (Type: {type(student.max_price)})")
#     print(f"Preferred Halls: {student.preferred_hall_ids} (Type: {type(student.preferred_hall_ids)})")
#     print(f"Roommate Prefs: {student.roommate_preferences} (Type: {type(student.roommate_preferences)})")
#     print(f"Accessibility: {student.accessibility_need} (Type: {type(student.accessibility_need)})")
#     print(f"Residence Type: {student.preferred_residence_type} (Type: {type(student.preferred_residence_type)})")
#     print("-" * 50)
```
